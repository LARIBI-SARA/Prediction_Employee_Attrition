```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

##Logistic Regression


```python
##Model
from sklearn.linear_model import LogisticRegression as LR
classifier = LR(random_state = 42).fit(Xtrain,ytrain)
y_predict = classifier.predict(Xtrain)
y_predict
```

```python
def accuracy(y,y_predict):
  acc = 0 #Le nombre d'Ã©chantillons correctement prÃ©dit
  for index in range(len(y)):
    if y[index] == y_predict[index]:
      acc +=1
  return acc/len(y)*100 ## Pourcentage d'Ã©chantillons correctement prÃ©dits

train_accuracy = accuracy(ytrain,y_predict)
print('Train accuracy : {:2f} %'.format(train_accuracy))
```

```python
y_test_pred = classifier.predict(Xtest)
test_accuracy = accuracy(ytest,y_test_pred)
print('Test accuracy : {:2f} %'.format(test_accuracy))
```
